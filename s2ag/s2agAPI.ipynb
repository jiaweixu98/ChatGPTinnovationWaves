```python
# https://github.com/allenai/s2-folks/blob/main/examples/Webinar%20Code%20Examples/PaperSearch.py
# Can only process 500 paper ids at a time.
# Can only return up to 10 MB of data at a time.
# 1 request per second for the following endpoints:/paper/batch; /paper/search; /recommendations
# 10 1 request per second for the following endpoints for others
import requests
import json
import os
```

```python
# Fetch API key from environment variables
api_key = 'jUP8YPo5KBaTVnjxhlpNy3dgkLBk37jxaZhyAWuU'

# Set up the headers with the API key
headers = {
    "x-api-key": api_key
}

# Make the POST request with headers
arxiv_id_queries = ['ARXIV:2005.14165','2310.00194']

# paper

r = requests.post(
    'https://api.semanticscholar.org/graph/v1/paper/batch',
    headers=headers,
    json={
        "ids": arxiv_id_queries
    },
    params={'fields': 'referenceCount,citationCount,title,venue,authors,s2FieldsOfStudy,publicationDate,journal,publicationTypes,embedding,tldr'},

)

# Check if the request was successful
if r.status_code == 200:
    print(json.dumps(r.json(), indent=2))
else:
    print(f"Error: {r.status_code}")
    print(json.dumps(r.json(), indent=2))
```
